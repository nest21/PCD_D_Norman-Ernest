```python
import cv2
from matplotlib import pyplot as plt
```

```python
# Fungsi untuk menampilkan gambar menggunakan Matplotlib
def show_image(title, image, cmap=None):
    plt.figure(figsize=(10, 10))
    plt.imshow(image, cmap=cmap)
    plt.title(title)
    plt.axis('off')
    plt.show()
```

```python
# Baca gambar dalam format BGR
image_path = 'bmw.jpeg'
image = cv2.imread(image_path)
```

```python
# Pastikan gambar terbaca
if image is None:
    print("Gambar tidak ditemukan atau tidak dapat dibaca.")
else:
    # Konversi BGR ke Grayscale
    gray_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
    
    # Konversi BGR ke HSV
    hsv_image = cv2.cvtColor(image, cv2.COLOR_BGR2HSV)

    # Tampilkan hasil konversi
    show_image('Original Image', cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
    show_image('Grayscale Image', gray_image, cmap='gray')
    show_image('HSV Image', cv2.cvtColor(hsv_image, cv2.COLOR_HSV2RGB))

    # Simpan hasil konversi ke file
    cv2.imwrite('gray_image.jpg', gray_image)
    cv2.imwrite('hsv_image.jpg', hsv_image)
```
