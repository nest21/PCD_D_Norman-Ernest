```python
import cv2
import numpy as np
from matplotlib import pyplot as plt
```

```python
# Fungsi untuk menampilkan gambar menggunakan Matplotlib
def show_image(title, image, cmap=None):
    plt.figure(figsize=(10, 10))
    plt.imshow(image, cmap=cmap)
    plt.title(title)
    plt.axis('off')
    plt.show()
```

```python
# Baca gambar dalam format BGR
image_path = 'bmw.jpeg'
image = cv2.imread(image_path)
```

```python
# Pastikan gambar terbaca
if image is None:
    print("Gambar tidak ditemukan atau tidak dapat dibaca.")
else:
    # Konversi gambar dari BGR ke RGB untuk ditampilkan dengan benar menggunakan Matplotlib
    image_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

    # 1. Tingkatkan kecerahan gambar
    brightness = 50
    bright_image = cv2.convertScaleAbs(image, alpha=1, beta=brightness)

    # 2. Filter Median
    median_filtered_image = cv2.medianBlur(image, 5)

    # 3. Filter Rata-rata
    kernel_size = (5, 5)
    mean_filtered_image = cv2.blur(image, kernel_size)

    # Tampilkan hasil konversi
    show_image('Original Image', image_rgb)
    show_image('Brightened Image', cv2.cvtColor(bright_image, cv2.COLOR_BGR2RGB))
    show_image('Median Filtered Image', cv2.cvtColor(median_filtered_image, cv2.COLOR_BGR2RGB))
    show_image('Mean Filtered Image', cv2.cvtColor(mean_filtered_image, cv2.COLOR_BGR2RGB))

    # Simpan hasil konversi ke file
    cv2.imwrite('/mnt/data/bright_image.jpg', bright_image)
    cv2.imwrite('/mnt/data/median_filtered_image.jpg', median_filtered_image)
    cv2.imwrite('/mnt/data/mean_filtered_image.jpg', mean_filtered_image)
```
